```python
#Load Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#reading the dataset
df = pd.read_csv ('./Absenteeism.csv')
```

```python
df
```

```python
#Create new Dataset without the ID column
df1 = df.drop('ID', axis=1)
df1
```

```python
#Show Key Statistics
df1.describe()
```

```python
#Correct the data in column 'Pet'
df1.loc[df1["Pet"] > 1, "Pet"] = 1
df1
```

```python
#Show Key Statistics
df1.describe()
```

```python
#Check columns in dataset
df1.columns
```

```python
#Boxplot Visualization
plt.figure(figsize=(15,10))
sns.boxplot(data=df1)
```

```python
#Scatterplots - Absenteeism vs. Independent Variables

plt.figure(figsize = (15, 20))

cnt = 1
for col in df1.columns:
    if (col == 'Absenteeism'):
        continue
    
    plt.subplot(5, 2, cnt)
    lmgraph = sns.regplot(x = df1[col], y = df1.Absenteeism)
    lmgraph.set_xlabel(col)
    lmgraph.set_ylabel('Absenteeism')
    cnt += 1    
```

```python
#Define x and y variable
x = df1.drop('Absenteeism',axis=1).to_numpy()
y = df1['Absenteeism'].to_numpy()

#Create Train and Test Datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=200)

#Scale the Data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train2 = sc.fit_transform(x_train)
x_test2 = sc.transform(x_test)
```

```python
#Script for Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn import metrics  

for name,method in [('Linear Regression', LinearRegression(n_jobs=-1))]: 
    method.fit(x_train2,y_train)
    predict = method.predict(x_test2)

print('Method: {}'.format(name))   

#Coefficents
print('\nIntercept: {:0.2f}'.format(float(method.intercept_)))
coeff_table=pd.DataFrame(np.transpose(method.coef_),df1.drop('Absenteeism',axis=1).columns,columns=['Coefficients'])
print('\n')
print(coeff_table)
    
#MAE,MSE and RMSE
print('\nR2: {:0.2f}'.format(metrics.r2_score(y_test, predict)))
adjusted_r_squared2 = 1-(1-metrics.r2_score(y_test,predict))*(len(y)-1)/(len(y)-x.shape[1]-1)
print('Adj_R2: {:0.2f}'.format(adjusted_r_squared2))

print('\nMean Absolute Error: {:0.2f}'.format(metrics.mean_absolute_error(y_test, predict)))  
print('Mean Squared Error: {:0.2f}'.format(metrics.mean_squared_error(y_test, predict)))  
print('Root Mean Squared Error: {:0.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, predict))))
```
